Staffelter Hof Winery is Germany's oldest business, established in 862 under the Carolingian dynasty. It has continued to serve customers through dramatic changes in Europe, such as the Holy Roman Empire, the Ottoman Empire, and both world wars. What characteristics enable a business to stand the test of time?

To help answer this question, BusinessFinancing.co.uk researched the oldest company still in business in **almost** every country and compiled the results into several CSV files. This dataset has been cleaned.

Having useful information in different files is a common problem. While it's better to keep different types of data separate for data storage, you'll want all the data in one place for analysis. You'll use joining and data manipulation to work with this data and better understand the world's oldest businesses.

## The Data
`data/businesses.csv` and `data/new_businesses.csv`
|Column|Description|
|------|-----------|
|`business`|Name of the business (varchar)|
|`year_founded`|Year the business was founded (int)|
|`category_code`|Code for the business category (varchar)|
|`country_code`|ISO 3166-1 three-letter country code (char)|

`data/countries.csv`
|Column|Description|
|------|-----------|
|`country_code`|ISO 3166-1 three-letter country code (varchar)|
|`country`|Name of the country (varchar)|
|`continent`|Name of the continent the country exists in (varchar)|

`data/categories.csv`
|Column|Description|
|------|-----------|
|`category_code`|Code for the business category (varchar)|
|`category`|Description of the business category (varchar)|

```python
# Import necessary libraries
import pandas as pd

# Load the data
businesses = pd.read_csv("data/businesses.csv")
new_businesses = pd.read_csv("data/new_businesses.csv")
countries = pd.read_csv("data/countries.csv")
categories = pd.read_csv("data/categories.csv")
```

```python
# Which business categories are best suited to last many years, and on what continent are they?

# TODO:

# 1. Merge the 'businesses' DataFrame with the 'countries' DataFrame to get continent information.
columns = ["continent", "country", "business", "year_founded"]
businesses_country = businesses.merge(countries, on='country_code', how='left')

# 2. Select only the relevant columns: 'continent', 'country', 'business', and 'year_founded'.
country_b = businesses_country[columns]

# 3. Sort the data by 'year_founded' to ensure the oldest businesses come first.
# 4. Group the sorted data by 'continent' and select the first (oldest) business in each group.
# 5. Reset the index to return a clean DataFrame and display the result.
oldest_business_continent = (
    country_b.sort_values(by='year_founded')  
    .groupby('continent')                     
    .first()                                  
    .reset_index()                            
)
print(oldest_business_continent)
```

```python
# Which business categories are best suited to last many years, and on what continent are they?

# TODO:
# 1. Group the 'businesses' DataFrame by 'continent' and 'category'.
category_merged = businesses.merge(categories, how="inner", on="category_code")

# 2. For each group, find the minimum 'year_founded' (oldest business).
category_with_continent = category_merged.merge(countries, how="inner", on="country_code")

# 3. Create a new DataFrame called 'oldest_by_continent_category'.
grouped = category_with_continent.groupby(['continent', 'category'])

# 4. The new DataFrame should contain three columns: 'continent', 'category', and 'year_founded'.
oldest_by_continent_category = grouped.agg({'year_founded': 'min'}).reset_index()

# 5. Ensure the 'continent', 'category', and 'year_founded' columns are in the correct order.
oldest_by_continent_category = oldest_by_continent_category.sort_values(by=['continent', 'year_founded'], ascending=[True, True])

print(oldest_by_continent_category)
```
